```python
#Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#Read in data
df = pd.read_csv(r'C:\Users\jared\Downloads\movies.csv')
```

```python
df.head()
```

```python
#Check for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}% '.format(col, pct_missing))
```

```python
#Determine Column Data Types
df.dtypes
```

```python
#Sort by Gross
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
#Scatter Plot with Budget vs Gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget')
plt.show
```

```python
df.head()
```

```python
#Plot the Budget vs Gross using Seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
#Correlation
df.corr(method='pearson') #kendall,spearman
```

```python
#Heatmap of Correlation
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
#Categorical Varibles as Numbers
df_number = df
for col_name in df_number.columns:
    if(df_number[col_name].dtype == 'object'):
        df_number[col_name] = df_number[col_name].astype('category')
        df_number[col_name] = df_number[col_name].cat.codes
        
df_number.head()
```

```python
#Correlation with All Variables
correlation_matrix = df_number.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```
